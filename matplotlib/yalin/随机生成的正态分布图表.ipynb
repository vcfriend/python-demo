```python
import matplotlib.pyplot as plt
plt.style.use(plt.style.available[0])  #可以通过下标进行选择适合自己的画图样式
from matplotlib.colors import ListedColormap

from sklearn.datasets import make_classification
```

```python
plt.figure(1,figsize=(8,8)) # 创建第一个画板（figure）
'''有六个可选参数来控制子图布局。值均为0~1之间。其中left、bottom、right、top围成的区域就是子图的区域。wspace、hspace分别表示子图之间左右、上下的间距。实际的默认值由matplotlibrc文件控制的
'''
plt.subplots_adjust(bottom=.05, top=.95, left=.05, right=.95,wspace=.2, hspace=.2)

plt.subplot(221) # 第一个画板的第一个子图

# 一个信息特性，每个类一个集群
plt.title("One informative feature, one cluster per class", fontsize='small')
X1, Y1 = make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=1,
                             n_clusters_per_class=1, flip_y=0.5)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')


plt.subplot(222)

plt.title("Two informative features, one cluster per class", fontsize='small')
# 两个信息特性，每个类一个集群
X1, Y1 = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1, flip_y=0.5)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')

plt.subplot(223)
# 两个信息特性，每个类两个集群
plt.title("Two informative features, two clusters per class",
          fontsize='small')
X2, Y2 = make_classification(n_features=2, n_redundant=0, n_informative=2)
plt.scatter(X2[:, 0], X2[:, 1], marker='o', c=Y2,
            s=25, edgecolor='k')

plt.subplot(224)
# 多类，两个信息特性，一个集群
plt.title("Multi-class, two informative features, one cluster",
          fontsize='small')
X1, Y1 = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1, n_classes=3)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
```
